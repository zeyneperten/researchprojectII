```python
# load real data pickle file
import pickle
with open('data/rat944_sleep_center.pkl', 'rb') as f:
    data944 = pickle.load(f)  
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scripts.simulation.rank_correlation as rs
nrm = np.load("scripts/simulation/nrm.npy")
```

```python
# dictionary_keys(['exp_phases', 'immobility_durations', 'bursts', 'burst_phases', 'peaks_id', 'seqs', 'seq_method', 'seq_firstspike', 'removed_cells', 'tt_key'])
# len(data944['seqs']): 6660

seqs = data944['seqs']
```

```python
# compute rank correlation for each sequence plot correlation matrix

n_seq = len(seqs)
rc_matrix = np.zeros((n_seq, n_seq))
for i in range(n_seq):
    for j in range(i+1, n_seq):
        rc, ln = rs.rankseq(seqs[i], seqs[j])
        rc_matrix[i, j] = rc
        rc_matrix[j, i] = rc  # symmetric
        
plt.figure(figsize=(8, 6))
plt.imshow(rc_matrix, cmap='coolwarm', aspect='auto', vmin=-1, vmax=1)
plt.colorbar(label='Rank Correlation (rc)')
plt.title('Pairwise Rank Correlation Matrix')
plt.xlabel('Sequence Index')
plt.ylabel('Sequence Index')
print(f"Rank Correlation Matrix Shape:\n{rc_matrix.shape}")
print(f"Rank Correlation Matrix Summary:\nMin: {np.min(rc_matrix)}, Max: {np.max(rc_matrix)}, Mean: {np.mean(rc_matrix)}, Std: {np.std(rc_matrix)}")
plt.show()
```
